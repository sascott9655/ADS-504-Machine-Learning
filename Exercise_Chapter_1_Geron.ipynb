###1. How would you define machine learning?

###3. What is a labeled training set?

A labeled training set is a training set that has been categorized before the model learns the data, which means the outcome of that observation is known before the model trains the data.

###9. What is an online learning system?

A system that learns data incrementally as opposed to batch learning system. It is good at learning rapidly to both changing and autonomous data.

###10. What is out-of-core learning?

Out-of-core learning is when there is a lot of data that cannot fit on CPU memory and the out-of-core learning chops the data into mini-batches.

###11. What type of algorithm relies on a similarity measure to make predictions?

###12. What is the difference between a model parameter and a model hyperparameter?

A model parameter is a parameter that directly influences the prediction while the model hyperparameter needs to be known before training and affects the rate at which a model learns the data.

###15. If your model performs great on the training data but generalizes poorly to new instances, what is happening? Can you name three possible solutions?

What is probably happening is the data is learning the training set too well and that is a major indicator of overfitting. Possible solutions are having more training data, using a less complex model, and or reducing the noise via regularization techniques

###16. What is a test set, and why would you want to use it?

You would want to use a test set to validate your training set's accuracy to make sure that the data is not overfitting.


###17. What is the purpose of a validation test?

A validation test is used to compare models. It makes it possible to select the best model and hyperparameter.

###18. What is a train-dev set, when do you need it, and how do you use it?

A train dev set is a second training set for your rare training data, and you need it when you have training data that does not occur often and you use it to validate your main training set.

###19. What can go wrong if you tune hyperparameters using the test set?
